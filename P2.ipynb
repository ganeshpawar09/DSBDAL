```python
import matplotlib.pyplot as plt

columns=["math score","reading score","writing score"]
df.boxplot(column=columns)
plt.title("Boxplot of Math, Reading, and Writing Scores")
plt.show()
```

```python
print((df["math score"]<20).sum())
print((df["math score"]<30).sum())

print((df["writing score"]<20).sum())
print((df["writing score"]<30).sum())

print((df["reading score"]<20).sum())
print((df["reading score"]<30).sum())


```

```python
df=df[df["math score"]>30]
df=df[df["reading score"]>30]
df=df[df["writing score"]>30]
```

```python
columns=["math score","reading score","writing score"]
df.boxplot(column=columns)
plt.title("After Removing Outliers")
plt.show()
```

```python
df["math score"]=(df["math score"]-df["math score"].min())/(df["math score"].max()-df["math score"].min())
df["reading score"]=(df["reading score"]-df["reading score"].min())/(df["reading score"].max()-df["reading score"].min())
df["writing score"]=(df["writing score"]-df["writing score"].min())/(df["writing score"].max()-df["writing score"].min())
```
